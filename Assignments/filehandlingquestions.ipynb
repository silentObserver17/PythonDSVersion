### Assignment 5: Counting Words in a File

Write a function that reads the contents of a file named `document.txt` and returns the number of words in the file.

```python
def count_words():
    total_words = 0
    with open("sample_files/document.txt", "r") as file:
        for l in file:
            words = l.split()
            total_words += len(words)

    return total_words

print(count_words())
```

```python
## Pythonic solution:
def count_words():
    with open("sample_files/document.txt", "r") as file:
        return sum(len(line.split()) for line in file)

print(count_words())
```

### Assignment 8: Counting Lines, Words, and Characters

Write a function that reads the contents of a file named `stats.txt` and returns the number of lines, words, and characters in the file.

```python
def stats_file():
    with open("sample_files/stats.txt", "r") as f:
        lines = f.readlines()

    number_of_lines = len(lines)
    number_of_words = sum(len(line.split()) for line in lines)
    number_characters = sum(len(line) for line in lines)

    return (
        f"Number of lines: {number_of_lines}, "
        f"Number of words: {number_of_words}, "
        f"Number of characters: {number_characters}"
    )

stats_file()
```

### Assignment 9: Merging Multiple Files

Write a function that merges the contents of multiple files into a single file named `merged.txt`.


```python
def merge_files(file_paths):
    with open("sample_files/merged.txt", "w") as f:
        for file_path in file_paths:
            with open(file_path, "r") as in_file:
                for line in in_file:
                    f.write(line)

merge_files([
    "sample_files/sample.txt",
    "sample_files/source.txt",
    "sample_files/stats.txt",
])
```

### Assignment 10: Splitting a Large File

Write a function that splits a large file named `large.txt` into smaller files of 100 lines each.

```python
import math

def split_files():
    with open("sample_files/large.txt", "r") as f:
        lines = f.readlines()

    chunk_size = 100
    total_files = math.ceil(len(lines) / chunk_size)

    for i in range(total_files):
        start = i * chunk_size
        end = start + chunk_size
        chunk = lines[start:end]

        with open(f"sample_files/large_part_{i+1}.txt", "w") as out:
                out.writelines(chunk)

split_files()
```

### Assignment 11: Creating a Log File

Write a function that creates a log file named `activity.log` and writes log messages with timestamps.


```python
import datetime as dt

def logging(message):
    with open("sample_files/activity.log", "a") as f:
        f.write(dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S") + "|" + message + "\n")


logging("User logged in")
```

### Assignment 12: Binary File Operations

Write a function that reads a binary file named `image.bin` and writes its contents to another binary file named `copy_image.bin`.


```python
import os

# Define the file name and size (1 MB = 1024 * 1024 bytes)
file_name = "sample_files/image.bin"
file_size = 1024 * 1024

# Generate random bytes
random_data = os.urandom(file_size)

# Open the file in 'wb' mode (write binary)
with open(file_name, "wb") as file:
    file.write(random_data)

print(f"Successfully generated {file_name} with {file_size} bytes of random data.")
```

Write a function that reads a CSV file named `data.csv` and prints its contents as a list of dictionaries.

### Assignment 15: File Permission Handling

Write a function that attempts to read a file named `protected.txt` and handles any permission errors gracefully by printing an error message.

```python
def read_protected_file():
    try:
        with open("sample_files/protected.txt", "r") as file:
            print(file.read())
    except PermissionError:
        print("Error: Permission denied while accessing 'protected.txt")
    except FileNotFoundError:
        print("Error: File not found")

read_protected_file()
```
