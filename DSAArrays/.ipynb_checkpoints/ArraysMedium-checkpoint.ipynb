## Two Sum

#### 1. Brute Force Method(using two loops)

````python
def two_sum_brute(nums, target):
    n = len(nums)
    for i in range(n):
        for j in range(i+1, n):
            if nums[i] + nums[j] == target:
                return [i, j]

    return [-1,-1]

nums = [1, 6, 2, 10, 3]; target = 7
print(two_sum_brute(nums, target))
````

#### 2. Optimal Solution(Using Hashmap/Dictionary)

````python
def two_sum_optimal(nums, target):
    prefix_map = {}
    n = len(nums)

    for i in range(n):
        rem = target - nums[i]
        if rem in prefix_map:
            return [prefix_map[rem], i]

        prefix_map[nums[i]] = i

    return [-1,-1]

nums = [1, 6, 2, 10, 3]; target = 7
print(two_sum_brute(nums, target))
````

## Sort an array of 0's 1's and 2's

#### 1. Basic Approach

````python
def sort_zero_one_two(arr):
    zero_count = 0
    one_count = 0
    two_count = 0

    for num in arr:
        if num == 0:
            zero_count += 1
        elif num == 1:
            one_count += 1
        else:
            two_count += 1

    return [0] * zero_count + [1] * one_count + [2] * two_count

nums = [2,0,2,1,1,0]
print(sort_zero_one_two(nums))
````

````python
def sort_zero_one_two(arr):
    zero_count = arr.count(0)
    one_count = arr.count(1)
    two_count = arr.count(2)

    return ([0] * zero_count) + ([1] * one_count) + ([2] * two_count)

nums = [2,0,2,1,1,0]
print(sort_zero_one_two(nums))
````

#### 2. Using Dutch National Flag Algorithm

Maintain three pointers:
- low â†’ boundary for 0s
- mid â†’ current element
- high â†’ boundary for 2s
---
**Space Complexity:**  ```O(1)``` (it modifies the list "in-place" without creating new lists).

**Time Complexity:** ```O(n)``` (it only goes through the list once).

````python
def sort_colors(nums):
    low = 0
    mid = 0
    high = len(nums)-1

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1

nums = [2,0,2,1,1,0]
print(sort_zero_one_two(nums))
````

### Majority Element (>n/2 times)

#### 1. Using Dictionary to store Frequency

````python
def majority_element(nums):
    majority = {}
    n = len(nums)

    for num in nums:
        if num in majority:
            majority[num] += 1
        else:
            majority[num] = 1

    for key,value in majority.items():
        if value > n//2:
            return key

    return -1

nums = [2,2,1,1,1,2,2]
print(majority_element(nums))
````

- Cleaner Version


-   â± Time: **O(n)**

-   ðŸ’¾ Space: **O(n)**


````python
def majority_element(nums):
    majority = {}
    n = len(nums)

    for num in nums:
        majority[num] = majority.get(num, 0) + 1
        if majority[num] > n//2:
            return num

    return -1

nums = [2,2,1,1,1,2,2]
print(majority_element(nums))
````

#### 2. By Boyer-Moore Voting Algorithm

**Core Idea**
> If an element appears more than `n/2` times, it **cannot be completely canceled out** by other elements.

Think of it as **pair cancellation**:
-   Each time we see **different elements**, they cancel each other
-   Majority element survives because it appears **more than all others combined**

> **TC (Time Complexity):** ```O(n)``` We only iterate through the array once.

> **SC (Space Complexity):** ```O(1)``` We only store two variables (candidate and count), regardless of how large the input array is.


````python
def majority_element(nums):
    candidate = None
    count = 0

    for num in nums:
        if count == 0:
            candidate = num
        count += 1 if num == candidate else -1

    return candidate

nums = [2,2,1,1,1,2,2]
print(majority_element(nums))
````

## Kadaneâ€™s Algorithm â€” Maximum Subarray Sum

**Core Idea**
> If the sum so far becomes negative, it will only reduce the future sum â†’ discard it.

1.  Maintain:
    -   `current_sum` â†’ max subarray ending at current index
    -   `max_sum` â†’ global maximum

2.  For each element:
    -   `current_sum = max(num, current_sum + num)`
    -   `max_sum = max(max_sum, current_sum)`

````python
def kadane_algorithm(nums):
    current_sum = max_sum = nums[0]

    for num in nums[1::]:
        current_sum = max(current_sum + num, num)
        max_sum = max(max_sum, current_sum)

    return max_sum

nums = [-2,1,-3,4,-1,2,1,-5,4]
print(kadane_algorithm(nums))
````
